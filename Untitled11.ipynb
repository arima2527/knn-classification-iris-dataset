#Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x= df.drop(["target","flower_name"],axis="columns")
y= df.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)#random_state is used so that there is no biasness and we can randomly chooose elements after 42 counts and give it to train_test and 20% is given(0.2) therefore the cycle repeats itself and remaining data is with train
```

```python
len(x_train)
```

```python
len(x_test)
```

Create KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KNeighborsClassifier(n_neighbors=5)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
#plt.figure(figure=(10,5))
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.show()
```
